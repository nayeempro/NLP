Data Missing Detection and it's remedial measure

```python
import pandas as pd

# Sample DataFrame
data = {
    'id': [1, 2, None, 3, 5, 6, 6, None, 7],
    'score': [10, None, 20, 30, None, 50, 60, 60, 60],
    'roll': [10, 40, 20, 30, 60, 50, 60, 60, 60],
}
df = pd.DataFrame(data)
df.head()
```

```python
#Detecting Data Missing
null = df.isnull()
null.sum()
```

```python
null
```

```python
df.isnull().any(axis=1)
```

```python
#get row which contain null value
df[df.isnull().any(axis=1)]
```

```python
#checking percentage of missing value according to column
miss_percentage = df.isnull().mean() * 100
miss_percentage
```

```python
#checking custom null or empty string
dt = {
    'id': [1, 2, None, 3, 5, 6, 6, None, 7],
    'score': [10, None, 20, 30, None, 50, 60, 60, 60],
    'roll': [10, '', 20, 30, 60, 50, 60, 60, 60],
}
df1 = pd.DataFrame(dt)
df1

```

```python
import numpy as np
df1.replace("",np.nan,inplace=True)
df1
```

```python
df1[df1.isnull().any(axis=1)]
```

Missing Value Handling

```python
#Remove missing value if they are greater that 50%
clear_df = df.dropna() # remove the row which contain null value
clear_df
```

```python
#remove col wise
clear_df_colwise = df.dropna(axis=1)
clear_df_colwise
```

```python
#Imputation -- Fill with some thing
```

```python
df_word = df.fillna('unknown') # if we use inplace = True then it change original data set
df_word
```

```python
#Fill num value by mean median mode of column
df_score_fill = df['score'].fillna(df['score'].mean()) 
df_score_fill
```

```python
#Fill num value by  median  of column
df_score_fill_median = df['score'].fillna(df['score'].median()) 
df_score_fill_median
```

```python
df['score'].mode()[0]
```

```python
#Fill num value by  median  of column
df_score_fill_mode = df['score'].fillna(df['score'].mode()[0]) 
df_score_fill_mode
```

Propagation : filling missing value with ffill or bfill (forward & backword)

```python
df
```

```python
df_forward_fill = df.fillna(method='ffill') #forward fill
df_forward_fill
```

```python
#backward fill
df_backward_fill = df.fillna(method='bfill') #forward fill
df_backward_fill
```

Linear Interpolation

```python
df_linear_fill = df.interpolate(method='linear') 
df_linear_fill
```
